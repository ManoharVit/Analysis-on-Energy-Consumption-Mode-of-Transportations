```python
%%writefile dash_app.py

from dash import Dash, html, dcc
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

# df = pd.DataFrame(main_df, columns=["air", "highway", "transit", "railways","water"])
df = pd.read_csv('main_df.csv', index_col=0)


app = Dash(__name__, suppress_callback_exceptions=True)


app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label="Overall Comparison Tab", children=[
            html.Div([
                html.H3('Overall Comparison Tab'),
                html.Div(children=[
                    html.Label('Categories to plot'),
                    dcc.Dropdown(['air', 'highway', 'transit', 'railways', 'water'],
                                multi=True, id="dropdown"),
                    html.Br(),
                    dcc.Slider(1960, 2020, 5,
                        value=1980,
                        id='year-slider',
                        marks={i: '{}'.format(i) for i in range(1960,2020,5)},
                    )
            ], style={'padding': 10, 'flex': 1}),

            html.Div(children=[
                    html.Div(id='output_container_tab1', children=[]),
                    html.Br(),
                    dcc.Graph(id='tab1_graph', figure={})
                ], style={'padding': 10, 'flex': 1}),
            ])
        ]),
        dcc.Tab(label='Year Wise Comparison Tab', children=[
            html.Div(children=[
                html.H3('Year Wise Comparison Tab'),
                html.Br(),
                dcc.Slider(1960, 2020, 5,
                    value=2000,
                    id='bin-slider',
                    marks={i: '{}'.format(i) for i in range(1960,2021,5)},
                ),
                html.Br(),
                html.Div(id='output_container_tab2', children=[]),
                html.Br(),
                dcc.Graph(id="tab2_bar", figure={}, style={'display': 'inline-block'}),
                dcc.Graph(id="tab2_pie", figure={}, style={'display': 'inline-block'})
            ]),
        ]),
    ])
])

@app.callback(
    [Output(component_id='output_container_tab1', component_property='children'),
    Output(component_id='tab1_graph', component_property='figure'),
    Output(component_id='output_container_tab2', component_property='children'),
    Output(component_id='tab2_bar', component_property='figure'),
    Output(component_id='tab2_pie', component_property='figure')],
    [Input(component_id='dropdown', component_property='value'),
    Input(component_id='year-slider', component_property='value'),
    Input(component_id='bin-slider', component_property='value')]
)
def update_graph(dropdown_value, slider_value, bin_value):
    dff = df.copy()
    df_tab2 = df.copy()

    container = "The year chosen by user: {}".format(slider_value)
    filtered_dff = dff.loc[slider_value:]
    fig = px.line(filtered_dff, y=dropdown_value, title='Energy Consumed by each category during the chosen year to 2020')

    # Bar Chart Code
    container_tab2 = "The bar chart is showing the data regarding year: {}".format(bin_value)
    tab2_bar = px.bar(df_tab2.loc[bin_value], title="Energy Consumption in Selected Year", )


    # Pie Chart Code
    air_avg = df_tab2['air'].mean()
    highway_avg = df_tab2['highway'].mean()
    water_avg = df_tab2['water'].mean()
    transit_avg = df_tab2['transit'].mean()
    railways_avg = df_tab2['railways'].mean()
    avg_df = [air_avg, highway_avg, water_avg, transit_avg, railways_avg]
    tab2_pie = px.pie(avg_df, values=avg_df, names=df_tab2.columns[0:], title="Average Pollution Generated By Each Source over time")
    return container, fig, container_tab2, tab2_bar, tab2_pie

if __name__ == '__main__':
  app.run_server(debug=False)
```
